```python
import glob
import os
import pandas as pd
coins_data = pd.read_csv("coins_data.csv")
csv_files = glob.glob(os.path.join("history/", '*.csv'))
dfs = []

for file in csv_files:
    # Extract the coin name from the file name
    coin_name = os.path.splitext(os.path.basename(file))[0].split('_')[0]
    
    df = pd.read_csv(file, sep=';')
    coin_id = coins_data.loc[coins_data['name'] == coin_name, 'rank'].values[0]
    df['id'] = coin_id
    dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)
combined_df.to_csv('all_historical_data.csv', index=False)
```
